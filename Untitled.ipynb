```python
import pandas as pd
import numpy as np
df=pd.read_csv('SG.csv')
df = df.rename(columns={'SG类型':'SG type',
                            'T_g':'Tf',
                            'b or (1-n)':'b',
                            'τ':r'$\tau_0$',
                            'Mydosh parameter \n':'Mydosh parameter',
                            'VF _ tau_0':r'$\tau_{VF}$',
                            'VF _ Ea/KB':'Ea',
                            'Critial scaling _tau0':r'$\tau_{cs}$',
                            'Critial scaling _zmu':r'$z\nu$'})
         
df.info()
```

```python
#对数据进行可视化
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df
```

```python
#Canonical SG
df_1 = df[(df['SG type']=='Canonical') ]
df_1=df_1[['SG type','Mydosh parameter',r'$\tau_{VF}$',r'$\tau_{cs}$',r'$z\nu$']]

#df_1.describe()
df_1=df_1.fillna(df_1.mean())
df_1.describe()
```

```python
#Cluster SG
df_2 = df[(df['SG type']=='Cluster') ]
df_2=df_2[['SG type','Mydosh parameter',r'$\tau_{VF}$',r'$\tau_{cs}$',r'$z\nu$']]

df_2 = df_2.fillna(df_2.mean())
df_2.describe()
```

```python
df_all = pd.concat([df_1,df_2])
df_all[r'$\tau_{VF}$']=df_all[r'$\tau_{VF}$'].apply(lambda x : np.log10(x))
df_all[r'$\tau_{cs}$']=df_all[r'$\tau_{cs}$'].apply(lambda x : np.log10(x))
df_all
```

### ROC curves, Area-Under-Curve (AUC)
